# **Loan Default Prediction**

## **Executive Summary**

**Objective**:

The credit department wants to simplify the current loan approval decision making process for the home equity line of credit. The current process is manual and prone to human error and biases. In this project, the immediate objective is to build a classification model for predicting the probability of default for any given customer applications and provide the bank with recommendations for important features to consider when approving a loan. The model will adopt guidlines of the Equal Credit Opportunity Act to establish an empirically derived and statistically sound credit scoring framework. The model will be based on recent data from the existing underwriting process, including customer profile information and the loan performance of approved customers. It will be built using predictive modeling techniques, and it must be interpretable to provide customer justifications for adverse actions (i.e. application rejections). Other objectives would include gaining insights from the loan performance data to aid in the design of customer acquisition strategies and help the business to best utilize the model's predictions for decision-making.

**Key findings**:

This pool of loans had a high default rate at about 20%. Most approved applicants had established credit history with minimum prior delinquency records. The Debt-to-income ratio was already moderate to high. The loan was mostly intended to consolidate the existing debt. The average approved loan amount was $19K. This portfolio is risky and susceptible to the economic changes.

In terms of the feature importance for predicting default rate, the top three most important features are Debt-to-income ratio (DEBTINC), Age of the oldest credit line (CLAGE) and Number of delinquent credit lines (DELINQ). We can roughly segment the defaulters into three groups:
   - People with established credit history but with very high Debt-to-income ratio (ex. 44%+)
   - People with established credit history but with relatively higher number of delinquent credit lines
   - People with low Debt-to-income ratio but relatively new to credit

Although other features are not as powerful to predict the default rate, some of them could be used at the marketing stage to improve the prospect targeting. Some highlights are as following:
   - Customers with office jobs have the lowest default rate on average, while customers with sales related jobs or self-employed tend to have the highest default rate. The longer the duration of the present job, the lower the default rate.
   - More recent credit inquiries the customer has, the higher the default rate.
   - The lower the existing property value, the higher the default rate.
   - Home improvement loans tend to have higher default rate than the debt consolidation loans on average.

**Model specifications**:

We built three different classification models: logistic regression, decision tree classifier and random forest classifier. The performance ranking is **random forest classifier (Best), logistic regression (Middle) and decision tree classifier (Worst)**. Given that the financial impact of lending to a defaulter is much more severe than declining a potential good customer, the main metric used for comparing the model performance is how accurately model can predict defaulters (i.e. class 1 recall score), while balancing the overall accuracy to avoid declining too many potential good customers. 

The random forest model outperforms the other two models. It can capture 80% of true defaulters (recall for class 1 is 0.8). Out of all the loans that are classified as defaulters by the model, 67% are true defaulters which is a decent precision score. The overall accuracy of the model is 88%. Although the logistic regression model has higher recall for class 1 (0.9), its precision is only at 0.47. The overall accuracy score is 0.78 for the logistic regression model, lower than the random forest model. Similarly, the recall for class 1 is 0.84 for the tuned decision tree model with a very low precision score at 0.47.

We know the interpretability of model prediction is very important, SHAP force plot can be utilized to visualize the prediction outcome and explain the contribution of each feature to a particular prediction. By leveraging the SHAP force plot as well as the feature importance ranking, the decision making rationale of the random forest model can be explained relatively easily. 

**Recommendations and next steps**:

We recommend to implement the random forest model for the approval process. This automation can help to improve the profitability, increase operations efficiency, enhance customer experience and enable the business to grow faster along with the technology advancement. Model implementation may reduce default rate from 20% to 5%. It will be able to decline 80% of current defaulters, which will greatly improve the profitability of the product. For example, for the average loan amount with 15% APR and 5 year term, profit may increase by as much as 58%. Key assumptions are:

- Profit = Revenue from the interest collected - Loss due to Default - Operating Expense
- Average APR is 10%, term of the loan is 5 years
- Defaulter: Collect 10% of life of loan revenue, 90% loan amount default, operating expense is 15% of loan amount
- Non-defaulter: Collect all revenue, operating expense is 5% loan amount
- Under this scenario, model approval framework can increase business profitability by 58%.

We propose to take a cautious approach to change the current approval process via a champaign/challenger approach in order to ensure a smooth transition. We will engage the partner teams to obtain feedback and approval for the recommended solution and prioritize resources to deploy the model upon approval. In the short term, we recommend adjusting the current manual approval workflow with some quick wins while working on the model implementation.

There are some risks involved with this proposal mainly around random forest model interpretability and model performance degradation due to out of sample fitting issues (i.e. when the business experiences any changes internally or externally that are not reflected in the training dataset). ## **Data Description:**
The Home Equity dataset (HMEQ) contains baseline and loan performance information for 5,960 recent home equity loans. The target (BAD) is a binary variable that indicates whether an applicant has ultimately defaulted or has been severely delinquent. This adverse outcome occurred in 1,189 cases (20 percent). 12 input variables were registered for each applicant.


* **BAD:** 1 = Client defaulted on loan, 0 = loan repaid

* **LOAN:** Amount of loan approved.

* **MORTDUE:** Amount due on the existing mortgage.

* **VALUE:** Current value of the property. 

* **REASON:** Reason for the loan request. (HomeImp = home improvement, DebtCon= debt consolidation which means taking out a new loan to pay off other liabilities and consumer debts) 

* **JOB:** The type of job that loan applicant has such as manager, self, etc.

* **YOJ:** Years at present job.

* **DEROG:** Number of major derogatory reports (which indicates a serious delinquency or late payments). 

* **DELINQ:** Number of delinquent credit lines (a line of credit becomes delinquent when a borrower does not make the minimum required payments 30 to 60 days past the day on which the payments were due). 

* **CLAGE:** Age of the oldest credit line in months. 

* **NINQ:** Number of recent credit inquiries. 

* **CLNO:** Number of existing credit lines.

* **DEBTINC:** Debt-to-income ratio (all your monthly debt payments divided by your gross monthly income. This number is one way lenders measure your ability to manage the monthly payments to repay the money you plan to borrow.

```python
# Function to plot a boxplot and a histogram along the same scale

def histogram_boxplot(data, feature, figsize=(12, 5), kde=False, bins=None):
    """
    Boxplot and histogram combined

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (12,5))
    kde: whether to the show density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows = 2,      # Number of rows of the subplot grid = 2
        sharex = True,  # x-axis will be shared among all subplots
        gridspec_kw = {"height_ratios": (0.25, 0.75)},
        figsize = figsize,
    )                   # Creating the 2 subplots
    sns.boxplot(data = data, x = feature, ax = ax_box2, showmeans = True, color = "violet"
    )                   # Boxplot will be created and a star will indicate the mean value of the column
    sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2, bins = bins, palette = "winter"
    ) if bins else sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2
    )                   # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color = "green", linestyle = "--"
    )                   # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color = "orange", linestyle = "-"
    )                   # Add median to the histogram
```

## Treating Outliers

One way of handling the outlier is to add floor and cap using the IQR method. Calculate the threshold for outliers.

```python
#IQR can be calculated for all numerical variables except for DEROG, DELINQ, NINQ where IQR is zero.
num_cols2=['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'CLAGE',  'CLNO', 'DEBTINC']
```

```python
#Calcualte IQR = Q3 - Q1
#lower_whisker = Q1 - 1.5*IQR
#upper_whisker = Q3 + 1.5*IQR

Q1=df[num_cols].quantile(0.25)
Q3=df[num_cols].quantile(0.75)
IQR = Q3 - Q1
lower_whisker = Q1 - 1.5*IQR
upper_whisker = Q3 + 1.5*IQR
```

```python
#Combine quantile value and upper whisker together
x=pd.concat([df[num_cols].quantile(0.25), df[num_cols].quantile(0.5), df[num_cols].quantile(0.75), upper_whisker, df[num_cols].max(), (df[num_cols]>upper_whisker).sum()/df.shape[0]*100], axis=1)
x=x.rename(columns={0.25: 'Q1', 0.5: 'Q2', 0.75: 'Q3', 0:'upper_whisker', 1: 'max', 2: 'pct'})
x
```

Observations:
- The max value for LOAN, MORTDUE, VALUE, DEROG, DELINQ, NINQ all look real, so we will not cap them at upper_whisker.
- The max value for For YOJ, CLAGE, CLNO, DEBTINC all look too high, so we will take a closer look. We will check default rate for quantile 1, 2, 3 and values higher than upper_whisker. The intent is to see if large values higher than upper_whisker has different default rate. If so, we may want to keep the those large values and not to treat them as outliers.

```python
#Define a function to check default rate by bins for YOJ, CLAGE, CLNO, DEBTINC
def outlier_bin(df, var, x):
    bins=x.loc[var, :]
    df2=df[[var, 'BAD']]
    df2['bin']=pd.cut(df2[var], bins)
    print(var, pd.concat([df2.groupby(['bin'])['BAD'].mean(), df2['bin'].value_counts()], axis=1))
    print('*'*50)
```

```python
outlier_bin(df, 'YOJ', x[['Q1', 'Q2', 'Q3', 'upper_whisker', 'max']])
outlier_bin(df, 'CLAGE', x[['Q1', 'Q2', 'Q3', 'upper_whisker', 'max']])
outlier_bin(df, 'CLNO', x[['Q1', 'Q2', 'Q3', 'upper_whisker', 'max']])
outlier_bin(df, 'DEBTINC', x[['Q1', 'Q2', 'Q3', 'upper_whisker', 'max']])
outlier_bin(df, 'NINQ', x[['Q1', 'Q2', 'Q3', 'upper_whisker', 'max']])
```

The next step is to build a simple regression model and obtain the formula to impute MORTDUE and VALUE.

```python
# Importing libraries for building linear regression model
import statsmodels.api as sm

df2=df[['VALUE', 'MORTDUE']]
df2.dropna(inplace=True)

train_features=df2['VALUE']
train_target=df2['MORTDUE']
# Adding the intercept term
train_features = sm.add_constant(train_features)

# Calling the OLS algorithm on the train features and the target variable
ols_model_0 = sm.OLS(train_target, train_features)

# Fitting the Model
ols_res_0 = ols_model_0.fit()

print(ols_res_0.summary())
```

```python
#fill in missing values for MORTDUE AND VALUE using the linear regression results above
# MORTDUE=-1280+0.7147*VALUE
#VALUE=(MORTDUE+1280)/0.7147

for i in range(0,df.shape[0]):
    if pd.isnull(df.at[i, 'VALUE'])==1:
        df.loc[i, 'MORTDUE2']=df.loc[i, 'MORTDUE']
    elif pd.isnull(df.at[i, 'MORTDUE'])==0:
        df.loc[i, 'MORTDUE2']=df.loc[i, 'MORTDUE']
    else:
        df.loc[i, 'MORTDUE2']= min(-1280 + 0.714*df.loc[i, 'VALUE'], 399550) #cap the max 
```

### **Decision Tree - Hyperparameter Tuning**

* Hyperparameter tuning is tricky in the sense that **there is no direct way to calculate how a change in the hyperparameter value will reduce the loss of your model**, so we usually resort to experimentation. We'll use Grid search to perform hyperparameter tuning.
* **Grid search is a tuning technique that attempts to compute the optimum values of hyperparameters.** 
* **It is an exhaustive search** that is performed on the specific parameter values of a model.
* The parameters of the estimator/model used to apply these methods are **optimized by cross-validated grid-search** over a parameter grid.

**Criterion {gini, entropy}**

The function to measure the quality of a split. Supported criteria are gini for the Gini impurity and entropy for the information gain.

**max_depth** 

The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

**min_samples_leaf**

The minimum number of samples is required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. This may have the effect of smoothing the model, especially in regression.

You can learn about more Hyperpapameters on this link and try to tune them. 

https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html


```python
# Initialize a new decision tree classifier
dtree_estimator = DecisionTreeClassifier(class_weight = {0: w0, 1: w1}, random_state = 1)

# Set the grid of parameters to choose from. Please note the hyperparameters in the code below are chosen after a few trials.
parameters = {'max_depth': np.arange(2, 7),
              'criterion': ['gini', 'entropy'],
              'min_samples_leaf': [3, 4, 5, 10, 15]
             }

# Set recall score of class 1 as the scoring used to compare parameter combinations
scorer = metrics.make_scorer(recall_score, pos_label = 1)

# Run the grid search with 10 fold cross-validation
gridCV = GridSearchCV(dtree_estimator, parameters, scoring = scorer, cv = 10) 

# Fitting the grid search on the train data
gridCV = gridCV.fit(x_train, y_train)

# Set the classifier to the best combination of parameters
dtree_estimator = gridCV.best_estimator_

# Fit the best estimator to the data
dtree_estimator.fit(x_train, y_train)
```

```python
# Check what results are available to view from the GridSearchCV. 
# We want to print the hyperparameter values and mean_test_score and decide how to revise the hyperparameter values.

gridCV_results_dt=pd.DataFrame(gridCV.cv_results_)
gridCV_results_dt.columns
```

```python
# Print the gridSearchCV results
gridCV_results_dt[['param_criterion', 'param_max_depth', 'param_min_samples_leaf', 'mean_test_score']].sort_values('mean_test_score', ascending=False).head(10)
```

Observations: hyperparameters that resulted in the best mean_test_score are within the parameter value search range, so we will adopt these values.
